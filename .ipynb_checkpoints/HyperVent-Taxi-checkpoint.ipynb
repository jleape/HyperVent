```python
# preview trips data
trips_df = pd.read_csv("trips_2019-03.csv", nrows=10)
trips_df.head()
```

```python
# import data -> df
trips_df = pd.read_csv("trips_2019-03.csv", parse_dates=['tpep_pickup_datetime','tpep_dropoff_datetime'],
                      usecols=[1,2,3,7,8])
```

```python
# pick-up and drop-off times
trips_df['pu_hr'] = [x.hour + round(x.minute / 60,1) for x in trips_df['tpep_pickup_datetime']]
trips_df['do_hr'] = [x.hour + round(x.minute / 60,1) for x in trips_df['tpep_dropoff_datetime']]

# pick-up and drop-off day types
trips_df['pu_day'] = ['weekday' if x.weekday() < 5 else 'weekend' for x in trips_df['tpep_pickup_datetime']]
trips_df['do_day'] = ['weekday' if x.weekday() < 5 else 'weekend' for x in trips_df['tpep_dropoff_datetime']]
```

```python
# aggregate trips by taxi zone, day and time
pu_sum = trips_df.groupby(['PULocationID','pu_day','pu_hr'])[['passenger_count']].agg('sum').reset_index()
do_sum = trips_df.groupby(['DOLocationID','do_day','do_hr'])[['passenger_count']].agg('sum').reset_index()
```

```python
# join to calculate flux
flux_df = pd.merge(pu_sum,do_sum,how='outer',
                   left_on=['PULocationID','pu_day','pu_hr'],
                   right_on=['DOLocationID','do_day','do_hr'])

flux_df.fillna(0,inplace=True)

flux_df['flux'] = flux_df['passenger_count_y'] - flux_df['passenger_count_x']

flux_df.drop(labels=['DOLocationID','do_day','do_hr','passenger_count_x','passenger_count_y'],axis=1,inplace=True)
flux_df.columns = ['zone','day','hr','flux']
```

```python
# calculate count of each day type
import collections
dates = set([x.date() for x in trips_df['tpep_pickup_datetime']])
day_types = ['weekday' if x.weekday() < 5 else 'weekend' for x in dates]
day_type_counts = collections.Counter(day_types)

# average flux by day type
flux_df['flux'] = flux_df['flux'] / [day_type_counts[x] for x in flux_df['day']]
```

```python
def plot_cartogram(day, hr):
    '''
    Plots cartogram according to average flux at hour of given day type
    hr must be a float between 0 and 24 inclusive
    day must be in {'weekday','weekend'}
    '''

    # filter flux_df
    flux_df[flux_df['day'] == day & flux_df['hr'] == hr]
    
    # join zones with flux
    taxi_zones.merge()
    # set figure
    fig, ax = plt.subplots(1, 1)
    fig.set_size_inches(15, 15)
    fig.set_dpi(100)
    plt.ioff()
    
    # plot cartogram
    carto = make_cartogram(taxi_zones, 'ppl', 2, inplace=False);
    nyc = carto.plot(column='ppl',cmap='cividis',ax=ax);
    nyc.axis("off");
        
    file_name = os.path.join("anim", str(f"hr_{hr}.png"));
    plt.savefig(file_name, bbox_inches="tight", pad_inches=0.1);
    plt.close();
                             
    return file_name
```

```python
# test the function
plot_cartogram(8)
```

```python
# array of times
hrs = np.linspace(0,24,97).tolist()
```

```python
image_files = []
```

```python
for hr in hrs:
    file_name = plot_cartogram(hr);
    image_files.append(file_name);
```

```python
images = []

for file_name in image_files:
    images.append(imageio.imread(file_name))

gif_path = "movie.gif"
imageio.mimsave(gif_path, images, fps=24)
```
